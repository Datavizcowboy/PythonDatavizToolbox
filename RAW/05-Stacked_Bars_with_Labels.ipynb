```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
path='./DATA/'
fronts = path+'Data_Veg_Delta_Stacked.txt'
fronts2 = path+'Data_Soil_Delta_Stacked.txt'
```

```python
data0 = []
data2 = []

for i in range(0,14):

 name=[]
 avg=[]
 them = []
 fp = open(fronts, 'r')
 for g in fp.readlines()[1:]:
    model_name=g.strip()
    data=model_name.split('\t')
    avg.append(float(data[i+1]))
    for cc in range(0,1):
      name.append(data[0])

 data0.append(avg)

name.append("Model Mean")

for i in range(0,14):
 name2=[]
 avg2=[]
 fs = open(fronts2, 'r')
 for s in fs.readlines()[1:]:
    model_name=s.strip()
    data=model_name.split('\t')
    avg2.append(float(data[i+1]))
 data2.append(avg2)

fr = open(fronts2, 'r')

for r in fr.readlines()[:1]:
    model_name=r.strip()
    datareg=model_name.split('\t')

regions =  datareg[1:]
```

```python
data_t = np.array(data0)
datas_t = np.array(data2)

data = np.vstack((data_t.T, np.mean(data_t, axis=1)))
datas = np.vstack((datas_t.T, np.mean(datas_t, axis=1)))

name = name + name

label_top = [sum(x) for x in zip(data.T, datas.T)]
```

```python
length = len(data)
width = 0.8 # width of bar
x = np.arange(length)
xti = np.arange(0.5,14.5,1)
```

```python
plt.rcParams['font.family'] = "Arial"
plt.rcParams['font.size'] = "18"

# fig=plt.figure(figsize=(11,5))
fig, ax = plt.subplots(1,1)

total = [np.int(sum(x)) for x in zip(data[:,0],datas[:,0])]
total = total + total

for k in range(0,1):
  ax.bar(x + (k * (width+0.01)), data[:,0], width, color='#F3BC81', ecolor='k')
  ax.bar(x + (k * (width+0.01)), datas[:,0], width, bottom=data[:,0], color='#7ECD14')

  rects = ax.patches
  labeltop = label_top[0]
  labels = name

  for rect, label in zip(rects, total):
    height = rect.get_height()
    if rect.get_y()!=0:
     height = rect.get_height()
     ax.text(
        rect.get_x() + rect.get_width() / 2, rect.get_y()+rect.get_height()+2, 
         label, ha="center", va="bottom",fontsize=10,weight='bold'
    )

  for rect, label in zip(rects, labels):
    height = rect.get_height()
    rect.set_edgecolor('black')
    if rect.get_y()!=0:
     ax.text(
        rect.get_x() + rect.get_width() / 2, rect.get_y()+rect.get_height()+100, 
         label, ha="center", va="bottom",fontsize=10, rotation=90
    )

  for rect in ax.patches:
    # Find where everything is located
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x()
    y = rect.get_y()

    # The height of the bar is the data value and can be used as the label
    label_text = f'{height:.0f}'  # f'{height:.2f}' to format decimal values

    # ax.text(x, y, text)
    label_x = x + width / 2
    label_y = y + height / 2

    # plot only when height is greater than specified value
    if height > 0:
        ax.text(label_x, label_y, label_text, ha='center', va='center', fontsize=8)

    ax.get_children()[21].set_color('#548412')
    ax.get_children()[10].set_color('#846749')
    ax.get_children()[10].set_edgecolor('black')
    ax.get_children()[21].set_edgecolor('black')

    ax.get_children()[23].set_color('red')
    ax.get_children()[24].set_color('red')
    ax.get_children()[25].set_color('red')
    ax.get_children()[26].set_color('red')
    ax.get_children()[27].set_color('red')

ax.set_ylim(0,1700,200)
ax.grid(linestyle='dotted',linewidth=.5, alpha=0.8)
ax.set_title(str(regions[0])+" BGC (PgC) - 4x$CO_2$",fontsize=14,loc='right')
ax.axes.get_xaxis().set_visible(False)
ax.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom=False)
```
