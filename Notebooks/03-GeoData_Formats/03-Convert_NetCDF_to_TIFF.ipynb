# Read a NetCDF file and convert it into TIFF format

In this notebook we read a NetCDF file of wind speed and we perform raster operations to create a TIFF file as an output

We first load useful libraries

```python
import xarray as xr 
import rioxarray as rio
import warnings
warnings.filterwarnings("ignore")
```

We specify the data path

```python
nc_file = xr.open_dataset('../../DATA/wind.nc')
```

The xarray we obtain is: 

```python
nc_file
```

We extract the variable of interest to be included into the output TIFF file

```python
variable = nc_file['UWND']
```

We now assign x and y dimensions based on the existing ones in the .nc file

```python
variable = variable.rio.set_spatial_dims(x_dim='FNOCXN79_64', y_dim='FNOCY')
```

With the variable information as well as the coordinate information we now create the .TIFF file 

```python
variable.rio.to_raster(r"../../OUTPUT/wind.tiff")
```
