# Read a NetCDF file, plot the contours and export them as GeoJSON

In this notebook a NetCDF file is read, then a variable is plotted as contour lines and these lines are saved as a GeoJSON

We first load useful libraries

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
from matplotlib.colors import rgb2hex
import geojson
from geojson import Feature, LineString, FeatureCollection
import warnings
warnings.filterwarnings("ignore")
```

We specify the data path and we open the file as an xarray

```python
data_path = '../../DATA/wind.nc'
wind = xr.open_dataset(data_path) 
```

We check the output of the xarray, where dimensions are specified with their native names in the file, the variable names and dimensions and the attributes assigned when creating the files. 

```python
wind
```

We define lat/lon and variables from the xarray

```python
lats=wind.FNOCY
lons=wind.FNOCXN79_64
variable=wind.UWND
```

We create a map with the variable displayed on a global scale using the contour( ) function from matplotlib. 

```python
fig, ax = plt.subplots(1,1)
contour = ax.contour(lons, lats, variable[0], levels=np.linspace(start=-15, stop=15, num=20), cmap=plt.cm.viridis)
```

We save the contour plots as a dictionary

```python
line_features = []
for collection in contour.collections:
    paths = collection.get_paths()
    color = collection.get_edgecolor()
    for path in paths:
        v = path.vertices
        coordinates = []
        for i in range(len(v)):
            lat = v[i][0]
            lon = v[i][1]
            coordinates.append((lat, lon))
        line = LineString(coordinates)
        properties = {
            "opacity": .5,
            "stroke-width": .5,
            "stroke": rgb2hex(color[0]),
        }
        line_features.append(Feature(geometry=line, properties=properties))
```

We diagnose the structure of the dictionary created

```python
line_features
```

We finally associate this dictionary to a GeoJSON file format

```python
feature_collection = FeatureCollection(line_features)
geojson_dump = geojson.dumps(feature_collection, sort_keys=True)
with open('../../OUTPUT/wind_contour_to.geojson', 'w') as fileout:
    fileout.write(geojson_dump)
```

The colors used for the different levels of the color palette can be extrated in hexadecimal format for its later use in Javascript visualization and legends
