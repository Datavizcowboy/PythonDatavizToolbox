# Read a NetCDF file and mask it with an SHP area

In this notebook we read a NetCDF file as an xarray of global wind speed data, we convert it into a geopandas dataframe and we use the clipping functions of geopandas to extract only the area defined by and SHP from the United States border.  

We first load useful libraries

```python
import xarray as xr
import pandas as pd
import geopandas as gp
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

We first read the NetCDF file as an xarray

```python
data_path = '../../DATA/wind.nc'
wind = xr.open_dataset(data_path)
```

The xarray is converted into a dataframe using the built-in function to_dataframe( )

```python
wind_df = wind.to_dataframe()
wind_df
```

This dataframe lacks an index to properly structure the data, so we reset it and add it. 

```python
wind_df = wind_df.reset_index()
wind_df
```

Now that the dataframe is structured, we convert it into a geopandas

```python
wind_geodf = gp.GeoDataFrame(wind_df, geometry=gp.points_from_xy(wind_df.FNOCXN79_64, wind_df.FNOCY))
wind_geodf
```

We read the target SHP area where we want to have the data clipped

```python
country = gp.read_file('../../DATA/cb_2018_us_nation_20m.shp')
```

The coordinate system has to be equal in order to operate between two georeferenced datasets

```python
wind_geodf.crs = country.crs
```

Finally we perform the clipping using the native geopandas clip( ) function

```python
mask = gp.clip(wind_geodf, country)
mask
```

The resulting geopandas is just 174 rows long compared to the original 10512 rows. Plotting the mask reveals the gridpoints extracted from the SHP file. 

```python
fig,ax = plt.subplots(1,1,dpi=144)
mask.plot(ax=ax, facecolor="none", edgecolor='grey');
```
