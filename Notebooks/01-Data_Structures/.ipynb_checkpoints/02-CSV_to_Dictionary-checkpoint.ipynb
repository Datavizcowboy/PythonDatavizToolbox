We first import the required modules

```python
import numpy as np
import pandas as pd
```

We then generate some random numbers

```python
data_path = '../../DATA/airports.csv'
```

And we create the scatter plot between the x and y data. 

```python
names=[]
lat=[]
lon=[]
city=[]
region=[]
country=[]
tuple_list = []
```

```python
fp = open(data_path, 'r')  

for g in fp.readlines()[1:]:
    model_name=g.strip()
    data=model_name.split(',')

    names.append(data[2])
    lat.append(data[3])
    lon.append(data[4])
    city.append(data[8])
    region.append(data[9])
    country.append(data[10])
    
fp.close()
```

```python
for i in range(0,len(names)):
    tuple_list.append([lat[i],lon[i]])

my_dict = [dict(zip(("lat", "lon"), x)) for x in tuple_list]

o_dict = {
 "name": "airports",
 "children": [
  {
   "name": "there",
   "children": my_dict}]}
```
