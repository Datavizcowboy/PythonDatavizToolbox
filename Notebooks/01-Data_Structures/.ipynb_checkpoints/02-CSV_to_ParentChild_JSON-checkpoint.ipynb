# Read CSV file and convert into JSON file with Parent/Child structure

We first load the necessary python libraries

```python
import json
```

Data is retrieved from the worldwide airport data stored in the DATA folder

```python
data_path = '../../DATA/airports.csv'
output_path = '../../DATA/'
```

Empty arrays are defined to store all the variables contained in the CSV file

```python
names=[]
lat=[]
lon=[]
city=[]
region=[]
country=[]
tuple_list = []
```

The CSV is read line by line, storing the data split by commas into the arrays

```python
fp = open(data_path, 'r')  

for g in fp.readlines()[1:]:
    line =g.strip()
    data=line.split(',')

    names.append(data[2])
    lat.append(data[3])
    lon.append(data[4])
    city.append(data[8])
    region.append(data[9])
    country.append(data[10])
    
fp.close()
```

```python
for i in range(0,len(names)):
    tuple_list.append([lat[i],lon[i]])

my_dict = [dict(zip(("lat", "lon"), x)) for x in tuple_list]

output_dict = {
 "name": "airports",
 "children": [
  {
   "name": "locations",
   "children": my_dict}]}
```

```python
output_dict
```

```python
with open(output_path+"airports_output_dictionary.json", "w") as f:
    json.dump(output_dict, f)
```
