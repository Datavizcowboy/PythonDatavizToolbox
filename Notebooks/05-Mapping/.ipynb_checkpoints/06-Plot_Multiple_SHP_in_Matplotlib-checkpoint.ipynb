# Read a multiple SHP file and plot each polygon in Matplotlib

In this notebook we open an SHP file with the perimeters of a wildfire that took place in Paradise, California. We iterate over the polygons to plot each one of them using matplotlib. 

We first load useful libraries

```python
import matplotlib.pyplot as plt
import shapefile
```

We read the SHP file using Shapefile

```python
data_path = '../../DATA/campfire.shp'
shape = shapefile.Reader(data_path)
```

We then extract a single shape by taking the first element of the shape 

```python
feature = shape.shapeRecords()[0]
first = feature.shape.__geo_interface__
first
```

We create nonetheless the matplotlib iterating over all the shape polygons

```python
fig,ax = plt.subplots(dpi=144)
for shape in shape.shapeRecords():
    x = [i[0] for i in shape.shape.points[:]]
    y = [i[1] for i in shape.shape.points[:]]
    ax.plot(x,y,linewidth=0.5)
plt.show()
```
