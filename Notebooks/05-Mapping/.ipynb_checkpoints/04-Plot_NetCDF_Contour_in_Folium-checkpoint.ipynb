# Plot NetCDF Contour in Folium

We first load useful libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import EngFormatter, StrMethodFormatter
import xarray as xr
import pandas as pd
import geopandas as gp
import datetime
from datetime import datetime
startTime = datetime.now()
import re
import folium
from folium import plugins

from cartopy.img_transform import warp_array
import cartopy.crs as ccrs
import matplotlib.cm
import geojsoncontour
# from geojson import Feature, LineString, FeatureCollection
# import geojson
from matplotlib.colors import rgb2hex
```

We then generate some random numbers

```python
d10_file = '../../DATA/wind.nc'
ds = xr.open_dataset(d10_file)
ds
```

Blablabla

```python
lats = ds.FNOCY
lons = ds.FNOCXN79_64
#var = ds.tephra_cloud_top[-1]
var = ds.UWND[0]
```

We then extract the coordinates and variables

```python
colors=['#ffffe5','#fff7bc','#fee391','#fec44f','#fe9929','#ec7014','#cc4c02','#8c2d04']
cvals  = [-15,-10,-5,0,5,10,15]
norm=plt.Normalize(min(cvals),max(cvals))
tuples = list(zip(map(norm,cvals), colors))
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", tuples)
```

```python
contourf = plt.contourf(lons,lats,var,levels=cvals,vmin=-15, vmax=15,alpha=1, cmap=cmap,norm=norm);
```

We create the plot

```python
geojson = geojsoncontour.contourf_to_geojson(
    contourf=contourf,
    min_angle_deg=3.0,
    ndigits=10,
    # opacity=1,
    stroke_width=1,
    fill_opacity=1)
geojson
```

```python
m = folium.Map(location=[lats.mean(), lons.mean()], zoom_start=7,tiles="cartodbdark_matter")
```

```python
g = folium.GeoJson(
    geojson,
    style_function=lambda x: {
        'color':     x['properties']['stroke'],
        'weight':    x['properties']['stroke-width'],
#        'weight':0    ,
        'fillColor': x['properties']['fill'],
        'fillOpacity': .6,
        'opacity':   1,
    })
g.add_to(m)
```

```python
col_neu = []

for i in range(cmap.N):
    rgba = cmap(i)
    # rgb2hex accepts rgb or rgba
    col_neu.append(matplotlib.colors.rgb2hex(rgba))

import branca.colormap as cmp
from branca.colormap import ColorMap, LinearColormap, StepColormap
from branca.element import MacroElement
from branca.element import Figure, JavascriptLink
from branca.utilities import legend_scaler
from branca.element import Element

step = cmp.StepColormap(
 colors,
 index=cvals,
 vmin=-15, vmax=15,
 )

e = Element("""
  var all = document.querySelectorAll('div.legend text')
  console.log(all)
  for(var i = 0; i < all.length; i++) {
   all[i].style.fill="white"
   all[i].style.fontSize=9
  }
  var ticks = document.querySelectorAll('div.legend g.tick text')
  console.log(ticks)
  for(var i = 0; i < ticks.length; i++) {
    var value = parseFloat(ticks[i].textContent.replace(',', ''))
    var newvalue = Math.pow(10.0, value).toFixed(1).toString()
    ticks[i].textContent = newvalue
    ticks[i].style.fill="white"
  }
""")
```

```python
step.add_to(m)

html = step.get_root()
html.script.get_root().render()
html.script.add_child(e)
```
