# Plot TIF in Matplotlib_IMSHOW

We first load useful libraries

```python
from osgeo import gdal
import matplotlib.pyplot as plt
import numpy as np
```

We open the dataset

```python
dataset = gdal.Open(r'../../DATA/chile_esa_cci.tif')

band2 = dataset.GetRasterBand(1) # Green channel
img1 = band2.ReadAsArray(0,0,dataset.RasterXSize, dataset.RasterYSize)
img = np.dstack(img1)
img1
```

We assign values to zero and flip the array upside down for accuracy

```python
plt.figure()
plt.imshow(img[0])
plt.savefig('../../OUTPUT/chile_matplotlib_imshow.png')
plt.show()
```
