# Read a NetCDF file as xarray, plot the contour lines and export via Folium

In this notebook we read a NetCDF file using Xarray, we create a contour plot using Matplotlib, we save the countour lines as a geojson dictionary, we project this geojson on a map using folium and we finally add a legend using the Branca package. The output is saved as an HTML file. 

We first load useful libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import xarray as xr
import pandas as pd
import geopandas as gp
import folium
from folium import plugins
import geojsoncontour
from matplotlib.colors import rgb2hex
```

We open the NetCDF file of global wind speed as an xarray element

```python
data_path = '../../DATA/wind.nc'
data = xr.open_dataset(data_path)
```

The variable 'data' contains the latitudinal and longitudinal information, the time dimension and the variable values. 

```python
data
```

We assign dedicated variables to the coordinates and the values

```python
lats = data.FNOCY
lons = data.FNOCXN79_64
wind = data.UWND[0]
```

We define the scale interval, colors and parameters for the values to be plotted

```python
colors=['#ffffe5','#fff7bc','#fee391','#fec44f','#fe9929','#ec7014','#cc4c02','#8c2d04']
cvals  = [-15,-10,-5,0,5,10,15]
norm = plt.Normalize(min(cvals),max(cvals))
tuples = list(zip(map(norm,cvals), colors))
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", tuples)
```

The tuples array associate values with colors 

```python
tuples
```

And the cmap creates the color scale for the values interval 

```python
cmap
```

Once the scales and colors are defined, we create the contour plot

```python
contourf = plt.contourf(lons,lats,wind,levels=cvals,vmin=-15, vmax=15,alpha=1, cmap=cmap,norm=norm);
```

The contour plot is projected into a geojson dictionary, containing all the coordinates of the curves and the colors associated with each contour line. 

```python
geojson = geojsoncontour.contourf_to_geojson(
    contourf=contourf,
    min_angle_deg=0.0,
    ndigits=10,
    # opacity=1,
    stroke_width=1,
    fill_opacity=1)
```

We set up a folium map using the dark mode cartoDB cartography, where we define the center point and the zoom level. 

```python
m = folium.Map(location=[lats.mean(), lons.mean()], zoom_start=3,tiles="cartodbdark_matter")
```

We now insert the geojson into the folium map, reading the color properties for each contour line. 

```python
g = folium.GeoJson(
    geojson,
    style_function=lambda x: {
        'color':     x['properties']['stroke'],
        'weight':    x['properties']['stroke-width'],
        'fillColor': x['properties']['fill'],
        'fillOpacity': .6,
        'opacity':   1,
    })
g.add_to(m)
```

The colors used in this plot are extracted, first with the .rgb2hex function from Matplotlib and then creating a colorbar using the Branca package. Formatting of the legend is done using javascript and HTML code. 

```python
col_neu = []

for i in range(cmap.N):
    rgba = cmap(i)
    # rgb2hex accepts rgb or rgba
    col_neu.append(matplotlib.colors.rgb2hex(rgba))

import branca.colormap as cmp
from branca.colormap import ColorMap, LinearColormap, StepColormap
from branca.element import MacroElement
from branca.element import Figure, JavascriptLink
from branca.utilities import legend_scaler
from branca.element import Element

step = cmp.StepColormap(
 colors,
 index=cvals,
 vmin=-15, vmax=15,
 )

e = Element("""
  var all = document.querySelectorAll('div.legend text')
  console.log(all)
  for(var i = 0; i < all.length; i++) {
   all[i].style.fill="white"
   all[i].style.fontSize=9
  }
  var ticks = document.querySelectorAll('div.legend g.tick text')
  console.log(ticks)
  for(var i = 0; i < ticks.length; i++) {
    var value = parseFloat(ticks[i].textContent.replace(',', ''))
    ticks[i].textContent = value
    ticks[i].style.fill="white"
  }
""")
```

The legend is then added into the folium map 

```python
step.add_to(m)

html = step.get_root()
html.script.get_root().render()
html.script.add_child(e)
```

Additionaly we can add landmarks into the map that can be explored with tooltips. To this end, we add the names and coordinates of certain volcanos as an example. 

```python
list_names=["Popocatepetl","Nevado del Ruiz","Cotopaxi", "Tungurahua","Pichincha","Merapi","Raung","Villarica","Fuego","Mt.Cameroon"]

list_catbond = [[19.27361, -98.70222], [4.883333, -75.31667], [-0.6847222, -78.43167], [-1.466944, -78.44167]
			, [-0.1772222, -78.59889], [-7.541389, 110.4461], [8.125, 114.0417], [-39.42, -71.93917], [14.47444, -90.88028]
			, [4.216667, 9.1725]]

for point in range(0, len(list_catbond)):
    folium.Marker(list_catbond[point],popup=list_names[point],icon=folium.Icon(color='lightgray', icon='fa-caret-up', prefix='fa')).add_to(m)
```

Finally we add a minimap in one corner of the map displayed for orientation purposes. 

```python
from folium import plugins
minimap = plugins.MiniMap()
m.add_child(minimap)
```

The output map is saved as an html file that can be opened in a local server. 

```python
m.save('../../OUTPUT/wind_folium.html')
```
