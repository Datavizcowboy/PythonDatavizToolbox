# Read a TIF file, select a raster band and plot it in Matplotlib

In this notebook we open a TIF file using GDAL, we select one of the raster bands and we create a simple plot using Matplotlib

We first load useful libraries

```python
from osgeo import gdal
import matplotlib.pyplot as plt
import numpy as np
```

We open the dataset

```python
dataset = gdal.Open(r'../../DATA/chile_esa_cci.tif')
```

From the dataset we select one raster band and we convert it into an array

```python
data = dataset.GetRasterBand(1).ReadAsArray()
```

The data is flipped upside down for accuracy

```python
data_flip = np.flipud(data) 
```

We plot the flipped variable using the matplotlib package

```python
fig,ax = plt.subplots(dpi=144)
ax.pcolormesh(data_flip)    
plt.show()
```
