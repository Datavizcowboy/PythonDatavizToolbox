# Plot NetCDF in Cartopy

We first load useful libraries

```python
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

We then generate some random numbers

```python
data = Dataset('../Data/Fire_2021_07_15.nc','r')
```

We then extract the coordinates and variables

```python
lons = data.variables['longitude'][:]
lats = data.variables['latitude'][:]
fwi=data.variables['fwi'][:,:]
data.close()
```

We create the plot

```python
ax = plt.axes(projection=ccrs.PlateCarree())
plt.contourf(lons, lats, fwi.data[0], 10,
             transform=ccrs.PlateCarree())
ax.coastlines()
plt.title('Cartopy Example')
plt.show()
```
