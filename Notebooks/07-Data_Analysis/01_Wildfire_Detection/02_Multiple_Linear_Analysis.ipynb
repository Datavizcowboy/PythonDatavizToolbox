Libraries we need are imported first

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import statsmodels.api as sm
import statsmodels.formula.api as smf 
import warnings
warnings.filterwarnings('ignore')
```

Data used in this exercise is stored in the DATA folder

```python
data = '../../../DATA/chile_fire_cluster_sequence_v2.json'
```

The dataframe is created using the .read_csv function of Pandas

```python
hotspot = pd.read_json(data)
# hotspot.info()
```

The original CSV file is structured as a dataframe with the corresponding categories

```python
# hotspot = hotspot[hotspot['mir_bt'].notna()]
# hotspot = hotspot[hotspot['tir_bt'].notna()]
hotspot = hotspot[hotspot['frp'].notna()]
# hotspot = hotspot[hotspot['frp_uncertainty'].notna()]
hotspot['Diff']=hotspot['mir_bt']-hotspot['tir_bt']
hotspot
```

After renaming the variable of the dataframe, we use .describe() to see essential statistics for each variable

## Regression

```python
# hotspot = hotspot[hotspot['satellite'].str.contains('AQUA')==True]
# hotspot = hotspot[hotspot['confidence'] > 0.9] 
```

```python
linear = sm.OLS(Y,X)
results = linear.fit()
results.summary()
```

```python
poly = smf.ols(formula='frp_uncertainty ~ 0 + I(Diff ** 3)', data=hotspot).fit()
poly.summary()
```

```python
x = pd.DataFrame({'Diff': np.linspace(hotspot.Diff.min(), hotspot.Diff.max(), 100)})
ax = hotspot.plot.scatter('Diff','frp',s=.5,color='k',alpha=0.2)
ax.plot(x.Diff, poly.predict(x), 'g-', label='Poly n=2 $R^2$=%.2f' % poly.rsquared,alpha=0.9)
# ax.set_ylim(0.1,1.1)
```
