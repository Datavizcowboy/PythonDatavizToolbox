## Sum per day

```python
hotspot['dayz']=pd.to_datetime(hotspot['detection_time']).dt.date
```

```python
hotspot['dayz']
```

```python
hotspot_dups = hotspot[hotspot.duplicated(['dayz'], keep=False)].copy()
hotspot_sum = hotspot_dups.groupby('dayz')["frp"].apply(lambda x : x.sum())
hotspot_sum = hotspot_sum.reset_index()
hotspot_sum
```

```python
fig, ax = pl.subplots(figsize=(12, 4)) 
timeseries = pl.scatter(hotspot_sum['dayz'], hotspot_sum['frp'],s=4)
```
