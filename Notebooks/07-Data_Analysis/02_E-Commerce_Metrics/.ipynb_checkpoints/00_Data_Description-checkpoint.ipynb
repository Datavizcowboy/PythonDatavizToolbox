Libraries we need are imported first

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import warnings
warnings.filterwarnings('ignore')
```

Data used in this exercise is stored in the DATA folder

```python
data_path = '../../../DATA/ecommerce_orig.txt'
```

The dataframe is created using the .read_csv function of Pandas

```python
data = pd.read_csv(data_path, delimiter = "\t")
```

The original CSV file is structured as a dataframe with the corresponding categories

```python
data
```

```python
data.info()
```

After renaming the variable of the dataframe, we use .describe() to see essential statistics for each variable

```python
data.describe()
```

```python
data = data[data['Revenue'] > 0] 
data
```

```python
dayz = data["weeknumberbyyear"].unique().tolist()
```

```python
performance_per_week = data.groupby(['weeknumberbyyear']).Visits.sum().reset_index()
revenue_per_week = data.groupby(['weeknumberbyyear']).Revenue.sum().reset_index()
orders_per_week = data.groupby(['weeknumberbyyear']).Orders.sum().reset_index()
```

```python
performance_per_week['Revenue']= revenue_per_week['Revenue'].values.tolist()
performance_per_week['Orders']= orders_per_week['Orders'].values.tolist()
```

```python
performance_per_week = performance_per_week[:-1]
```

```python
performance_per_week
```

## Quarters

```python
Q1_performance = performance_per_week[(performance_per_week.index>=0) & (performance_per_week.index<14)]
Q2_performance = performance_per_week[(performance_per_week.index>=14) & (performance_per_week.index<27)]
Q3_performance = performance_per_week[(performance_per_week.index>=27) & (performance_per_week.index<40)]
Q4_performance = performance_per_week[(performance_per_week.index>=40) & (performance_per_week.index<52)]
Q1_performance
```

```python
performance_per_week['Ratio_Rev_Ord']=performance_per_week['Revenue']/performance_per_week['Orders']
performance_per_week['Ratio_Ord_Vis']=performance_per_week['Orders']/performance_per_week['Visits']
performance_per_week.describe()
```

```python
X = performance_per_week['weeknumberbyyear']
Y = performance_per_week['Revenue']
fig, ax = pl.subplots(figsize=(8, 4)) 
gsd = pl.scatter(X, Y, s=4)
```
